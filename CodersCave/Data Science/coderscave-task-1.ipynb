```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/kaggle/input/gtd/globalterrorismdb_0718dist.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1', low_memory=False)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='iyear', data=data)
plt.title('Number of Terrorist Incidents per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='imonth', data=data)
plt.title('Number of Terrorist Incidents per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='country_txt', data=data, order=data['country_txt'].value_counts().index[:10])
plt.title('Top 10 Countries with Highest Terrorist Incidents')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
Killed_annual=data[['iyear','nkill']].groupby('iyear').sum()
Wounded_annual=data[['iyear','nwound']].groupby('iyear').sum()

fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)

Killed_annual.plot(kind="bar",color="red",figsize=(15,14),ax=ax0)
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("Year")
ax0.set_ylabel("Number of People Killed") 

Wounded_annual.plot(kind="bar",color="blue",figsize=(15,14),ax=ax1)
ax1.set_title("People Wounded in each Year")
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of People Wounded")

plt.show()

```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='extended', data=data)
plt.title('Extended vs. Non-extended Attacks')
plt.xlabel('Extended')
plt.ylabel('Count')
plt.show()
```

```python
Attacks=data['iyear'].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={'iyear':'years'})
plt.plot(Attacks['years'], Attacks['count'])
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title('Number of Attacks Over the Years')
plt.show()
```

```python
Terr_region=pd.crosstab(data.iyear,data.region_txt)
Terr_region.plot(kind="area", stacked=False, alpha=0.7,figsize=(20,10))
plt.title("Region wise attack",fontsize=18)
plt.xlabel("Year",fontsize=18)
plt.ylabel("Number of Attacks",fontsize=18)
plt.show()
```
